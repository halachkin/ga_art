```python
import os
import numpy as np
import pandas as pd
from scipy import misc
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc("savefig", dpi=300)
mpl.rc("font", size=3)
```

```python
def extract_generation_n_from_image_name(t):
    return int(t[t.find("img_") + 4:-4])

#return stats, params, images, where:
#stats  - list of dataframes
#params  - list of dataframes
#images - list of lists with tuples( n_generation, img) - each list of images corresponds to stats and params
def get_logs(read_every_n_image = 20):
    # scan all logs
    logs = []
    for root, dirs, files in os.walk(os.path.curdir):
        if root!= ".\\.ipynb_checkpoints" and files[0]!= "Untitled.ipynb": 
            logs.append( [os.path.join(root, ifile) for ifile in files] )
    # get stats
    stats = []
    for statsname in [filter(lambda x: "stats.csv" in x, log) for log in logs]:
        stats.append(pd.read_csv(statsname[0], sep = ';', index_col = False))
    # get params
    params = []
    for paramsname in [filter(lambda x: "params.csv" in x, log) for log in logs]:
        param = pd.read_csv(paramsname[0], sep = ':', index_col = False, header = -1) 
        param = param.transpose()
        param.columns = param.iloc[0]
        param.reindex(param.index.drop(0))
        params.append(param)
    # get images
    images = []
    for imgnames in [filter(lambda x: ".png" in x, log) for log in logs]:
        tmp = []
        imgnames = sorted(imgnames, key = lambda x: extract_generation_n_from_image_name(x))
        for i, imgname in enumerate(imgnames):
            if i % read_every_n_image == 0:
                tmp.append( (extract_generation_n_from_image_name(imgname), misc.imread(imgname)))
        images.append(tmp)
        
    return stats, params, images
        
```

```python
stats, params, images = get_logs()
```

```python
for idx, imgs in enumerate(images):
    plt.figure(idx)
    title =  ''.join([params[idx][column][0] + ":" + str(params[idx][column][1]) + '\n' for column in params[idx].columns[4:]])
    for i, img in enumerate(imgs[1::len(images[1])/3]):
        plt.subplot(1,10, i + 1)
        plt.imshow(img[1], cmap=plt.cm.gray,)
        if(i == 2):
            plt.title(title, fontsize = 2)
#         plt.title( "generation: " + str(img[0]), fontsize = 3)
        plt.axis("off")
print "generations:"  + str([img[0] for img in imgs[1::len(images[1])/3]])
```

```python
mpl.rc("savefig", dpi=150)
mpl.rc("font", size=5)
for idx, stat in enumerate(stats):
    plt.figure(idx, figsize = (5,2))
    title =  ''.join([params[idx][column][0] + ":" + str(params[idx][column][1]) + '\n' for column in params[idx].columns[4:]])
    plt.title(title)
    plt.xlabel("GENERATION")
    plt.ylabel("FITNESS")
    plt.plot(stat["GENERATION"], stat["FITNESS"])
    plt.grid()
```
